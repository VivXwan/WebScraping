```python
from selenium import webdriver
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
options = webdriver.ChromeOptions()

options.add_argument('--headless')
options.add_argument('--no-sandbox')

driver = webdriver.Chrome('chromedriver',options = options)
```

```python
import time
driver.get("https://summerofcode.withgoogle.com/archive/2021/organizations")
time.sleep(5)
```

```python
allOrgs = driver.page_source
```

```python
allOrgs
```

```python
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By
```

```python
gsocSoup = bs(allOrgs,"html.parser")
```

```python
gsocSoup.title
```

```python
nextButton = driver.find_element(By.XPATH, "/html/body/app-root/app-layout/mat-sidenav-container/mat-sidenav-content/div/div/main/app-organizations/app-orgs-grid/section[2]/div/div[2]/mat-paginator/div/div/div[2]/button[2]")
```

```python
import math
totalCompanies = int(gsocSoup.find("div",class_="mat-paginator-range-label").text.strip().split(" ")[-1])
totalPages = math.ceil(totalCompanies/50)
totalPages
```

```python
cards = driver.find_elements(By.CLASS_NAME, "card")
```

```python
name = []
url = []
for i in range(totalPages):
  allOrgs1 = driver.page_source
  gsocSoup1 = bs(allOrgs1,"html.parser")
  cards = gsocSoup1.findAll('div', class_="card")
  for card in cards:
    name.append(card.find('div',class_='name').text.strip())
    url.append("https://summerofcode.withgoogle.com"+card.a['href'])
  nextButton.click()
  time.sleep(1)
```

```python
len(name)
```

```python
import pandas as pd
```

```python
allCompanies = pd.DataFrame({"Name":name,"URL":url})
```

```python
allCompanies
```

```python
allCompanies.to_csv("GSoC_Companies.csv",index = True)
```
